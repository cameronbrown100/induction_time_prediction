```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def t_ind(P, S):
    A = 130.7
    B = 0.25
    V = 1e-5
    tg = 10000

    J = A * S * np.exp(-B/np.log(S)**2)
    t = np.round((np.log(1-P))/(-J*V),0)+tg
    return t
```

```python
def data_generator(S):
    P = np.random.rand()
    return t_ind(P, S)
```

```python
S_bounds = np.array([[1.4, 2]])
S_init = np.array([[1.4], [1.8]])
```

```python
S_vals = np.arange(S_bounds[:, 0], S_bounds[:, 1], 0.01).reshape(-1, 1)
S_init = np.array([[1.4], [1.8]])

t_true = t_ind(0.9, S_vals)

t_noise = []
for i in range(len(S_vals)):
  t_noise.append(data_generator(S_vals[i,0]))

t_init = [] 
for i in range(len(S_init)):
  t_init.append(data_generator(S_init[i,0]))

plt.plot(S_vals, t_true, 'y--', lw=2, label='Noise-free objective')
plt.plot(S_vals, t_noise, 'bx', lw=1, alpha=1, label='Noisy samples')
plt.plot(S_init, t_init, 'kx', mew=3, label='Initial samples')
plt.legend();
```
